Percobaan 1

```python
# TODO 0: Import library lain yang dibutuhkan
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
%matplotlib inline

# TODO 1: Lakukan load image pada variabel berikut
# hint: kalian bisa gunakan fungsi open()
gambarku = Image.open("/content/drive/MyDrive/Program Fungsional/download.png")

# TODO 2: Ubah gambar menjadi hitam-putih
# hint: kalian bisa gunakan fungsi convert()
gambarBW = gambarku.convert("L")

# TODO 3: Tambahkan text sesuai kriteria.
draw = ImageDraw.Draw(gambarBW)
direktoriFont = "/content/drive/MyDrive/Program Fungsional/Roboto-Bold.ttf"  # Ganti dengan direktori font yang sudah kamu unduh
font = ImageFont.truetype(direktoriFont, 24)
text = "M Zidan Rifardi_268"
text_width, text_height = draw.textsize(text, font)
text_x = (gambarku.width - text_width) // 2
text_y = 30
draw.text((text_x, text_y), text, font=font, fill=0)

# TODO 4: Simpan gambar hasil edit menggunakan fungsi save()
gambarBW.save("/content/drive/MyDrive/Program Fungsional/percobaan_1.jpg")

# TODO 5: Tampilkan hasil akhir gambar
img = Image.open("/content/drive/MyDrive/Program Fungsional/percobaan_1.jpg")
plt.imshow(img)
# plt.set_cmap('gray')
plt.show()
```

Percobaan 2

```python
# TODO 0: Import beberapa library lain yang dibutuhkan
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

# TODO 1: Buka gambar latar belakang (background) dan gambar yang ingin disisipkan (overlay)
background_image = Image.open("/content/drive/MyDrive/Program Fungsional/large-cat-breed-1553197454.jpg")
overlay_image = Image.open("/content/drive/MyDrive/Program Fungsional/download.png")  # Pastikan gambar overlay memiliki channel alpha (transparansi)

# TODO 2: Konversi overlay image ke mode RGB (menghilangkan alpha channel)
overlay_image = overlay_image.convert("RGBA")

# TODO 3: (optional) Perkecil ukuran gambar overlay menggunakan method resize()
overlay_image = overlay_image.resize((150, 100))

# Tentukan/Hitung koordinat tengah untuk menempatkan overlay
x_center = (background_image.width - overlay_image.width) // 2
y_center = (background_image.height - overlay_image.height) // 2

# TODO 4: Sisipkan gambar overlay ke dalam gambar background menggunakan method paste()
background_image.paste(overlay_image, (x_center, y_center), overlay_image)

# TODO 5: Simpan gambar hasil edit
background_image.save("/content/drive/MyDrive/Program Fungsional/percobaan_2.jpg")

# TODO 6: Tampilkan
img = Image.open("/content/drive/MyDrive/Program Fungsional/percobaan_2.jpg")
plt.imshow(img)
plt.show()
```

Percobaan 3

```python
from PIL import Image, ImageEnhance

# TODO 1: Buka gambar dengan Pillow
image = Image.open("/content/drive/MyDrive/Program Fungsional/large-cat-breed-1553197454.jpg")

# TODO 2: Ubah tingkah kecerahan (brightness) dan kontras (contrast)
enhancer = ImageEnhance.Brightness(image)
brightened_image = enhancer.enhance(1.5)

enhancer = ImageEnhance.Contrast(brightened_image)
final_image = enhancer.enhance(1.2)

# TODO 3: Simpan gambar hasil edit
final_image.save("/content/drive/MyDrive/Program Fungsional/percobaan_3.jpg")

# TODO 4: Tampilkan
img = Image.open("/content/drive/MyDrive/Program Fungsional/percobaan_3.jpg")
plt.imshow(img)
plt.show()
```
